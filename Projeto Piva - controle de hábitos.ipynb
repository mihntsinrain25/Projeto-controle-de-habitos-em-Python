```python
habitos = []

# Sumário

def mostrar_sumario():
    print("\n=== CONTROLE DE HÁBITOS ===")
    print("1 - Adicionar hábito")
    print("2 - Listar hábitos")
    print("3 - Remover hábito")
    print("4 - Ver percentual médio de cumprimento")
    print("5 - Atualizar frequência de um hábito")
    print("0 - Sair")
```

```python
def adicionar_habito(habitos):
    print("\n--- Adicionar novo hábito ---")
    nome = input("Nome do hábito: ")
    frequencia = int(input("Frequência atual (quantas vezes cumpriu na semana): "))
    meta = int(input("Meta semanal (ex: 7): "))

    habito = {
        "nome": nome,
        "frequencia": frequencia,
        "meta": meta
    }

    habitos.append(habito)
    print("Hábito adicionado com sucesso!")
```

```python
def listar_habitos(habitos):
    print("\n--- Lista de hábitos ---")
    if not habitos:
        print("Nenhum hábito cadastrado.")
    else:
        for i, h in enumerate(habitos):
            print(f"{i} - {h['nome']} | Frequência: {h['frequencia']} / Meta: {h['meta']}")
```

```python
def remover_habito(habitos):
    if not habitos:
        print("Não há hábitos para remover.")
        return

    print("\nHábitos cadastrados:")
    for i, h in enumerate(habitos):
        print(f"{i} - {h['nome']}")

    opcao = input("\nDigite o número do hábito que deseja apagar (ou apenas Enter para não apagar nada): ")

    if opcao != "":
        indice = int(opcao)
        if 0 <= indice < len(habitos):
            removido = habitos.pop(indice)
            print(f"Hábito '{removido['nome']}' foi removido.")
    else:
        print("Número inválido. Nenhum hábito foi removido.")
```

```python
def calcular_percentual_medio(habitos):
    print("\n--- Percentual médio de cumprimento ---")
    if not habitos:
        print("Nenhum hábito cadastrado.")
        return

    soma_percentuais = 0
    for h in habitos:
        if h["meta"] > 0:
            percentual = (h["frequencia"] / h["meta"]) * 100
            soma_percentuais += percentual

    media = soma_percentuais / len(habitos)4
    print(f"Percentual médio de cumprimento: {media:.2f}%")
```

```python
def atualizar_frequencia(habitos):
    print("\n--- Atualizar frequência de um hábito ---")
    if not habitos:
        print("Nenhum hábito cadastrado.")
        return

    # Mostrar hábitos com índice
    for i, h in enumerate(habitos):
        print(f"{i} - {h['nome']} | Frequência: {h['frequencia']} / Meta: {h['meta']}")

    indice = int(input("Digite o número do hábito que deseja atualizar: "))

    if 0 <= indice < len(habitos):
        incremento = int(input("Quanto deseja somar na frequência? "))
        habitos[indice]["frequencia"] += incremento
        print("Frequência atualizada com sucesso!")
    else:
        print("Índice inválido.")
```

```python
# Programa principal

while True:
    mostrar_sumario()
    opcao = input("Escolha uma opção: ")

    if opcao == "0":
        print("Saindo do programa...")
        break
    elif opcao == "1":
        adicionar_habito(habitos)
    elif opcao == "2":
        listar_habitos(habitos)
    elif opcao == "3": 
        remover_habito(habitos)
    elif opcao == "4":
        calcular_percentual_medio(habitos)
    elif opcao == "5":
        atualizar_frequencia(habitos)
    else:
        print("Opção inválida. Tente novamente.")
```
